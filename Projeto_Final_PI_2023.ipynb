```python
# Importar as bibliotecas do opencv
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# Caminho para a imagem (mude caso tente executar em sua m√°quina)
path = '/content/drive/MyDrive/PIM - 2023/Projeto/database/t1.jpg'

# Ler e printar a imagem
img_src = cv2.imread(path)
cv2_imshow(img_src)
```

```python
# Aplicar e printar a imagem com median blur com ksize = 21
blurred = cv2.medianBlur(img_src, 21)
cv2_imshow(blurred)
```

```python
# Aplica√ß√£o da mascara para detec√ß√£o de bordas na imagem sem blur
edgesx = cv2.Scharr(img_src, -1, dx=1, dy=0, scale=0.75,
                       delta=0, borderType=cv2.BORDER_DEFAULT)
edgesy = cv2.Scharr(img_src, -1, dx=0, dy=1, scale=0.75,
                       delta=0, borderType=cv2.BORDER_DEFAULT)
edges1 = edgesx + edgesy

# Print do resultado do processamento
cv2_imshow(edges1)


# Aplica√ß√£o da mascara para detec√ß√£o de bordas na imagem com blur
edgesx = cv2.Scharr(blurred, -1, dx=1, dy=0, scale=1,
                       delta=0, borderType=cv2.BORDER_DEFAULT)
edgesy = cv2.Scharr(blurred, -1, dx=0, dy=1, scale=0.75,
                       delta=0, borderType=cv2.BORDER_DEFAULT)
edges2 = edgesx + edgesy

# Print do resultado do processamento
cv2_imshow(edges2)
```
