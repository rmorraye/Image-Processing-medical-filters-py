```python
# Importar as bibliotecas do opencv
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from skimage.io import imread_collection

# Caminho para a imagem (mude caso tente executar em sua m√°quina)
path_1 = '/content/drive/MyDrive/1. Disciplinas Materias/2023.1 - PIM/Projeto/database/t1.jpg' # imagem 1
path_2 = '/content/drive/MyDrive/1. Disciplinas Materias/2023.1 - PIM/Projeto/database/t2.jpg' # imagem 2

# Ler e plotar
#   Imagem 1
img_src_1 = cv2.imread(path_1)
cv2_imshow(img_src_1)

#   Imagem 2
img_src_2 = cv2.imread(path_2)
cv2_imshow(img_src_2)
```

```python
# Aplicar e printar a imagem com median blur com ksize = 21
#   Imagem 1
blurred_1 = cv2.medianBlur(img_src_1, 21)
cv2_imshow(blurred_1)

#   Imagem 2
blurred_2 = cv2.medianBlur(img_src_2, 21)
cv2_imshow(blurred_2)
```

```python
#   Imagem 1
# Aplica√ß√£o da mascara para detec√ß√£o de bordas na imagem sem blur
edgesx_1 = cv2.Scharr(img_src_1, -1, dx=1, dy=0, scale=0.75,
                       delta=0, borderType=cv2.BORDER_DEFAULT)
edgesy_1 = cv2.Scharr(img_src_1, -1, dx=0, dy=1, scale=0.75,
                       delta=0, borderType=cv2.BORDER_DEFAULT)
edges1_1 = edgesx_1 + edgesy_1

# Print do resultado do processamento
cv2_imshow(edges1_1)


# Aplica√ß√£o da mascara para detec√ß√£o de bordas na imagem com blur
edgesx_1 = cv2.Scharr(blurred_1, -1, dx=1, dy=0, scale=1,
                       delta=0, borderType=cv2.BORDER_DEFAULT)
edgesy_1 = cv2.Scharr(blurred_1, -1, dx=0, dy=1, scale=0.75,
                       delta=0, borderType=cv2.BORDER_DEFAULT)
edges2_1 = edgesx_1 + edgesy_1

# Print do resultado do processamento
cv2_imshow(edges2_1)
```

```python
#   Imagem 2
# Aplica√ß√£o da mascara para detec√ß√£o de bordas na imagem sem blur
edgesx_2 = cv2.Scharr(img_src_2, -1, dx=1, dy=0, scale=0.75,
                       delta=0, borderType=cv2.BORDER_DEFAULT)
edgesy_2 = cv2.Scharr(img_src_2, -1, dx=0, dy=1, scale=0.75,
                       delta=0, borderType=cv2.BORDER_DEFAULT)
edges1_2 = edgesx_2 + edgesy_2

# Print do resultado do processamento
cv2_imshow(edges1_2)


# Aplica√ß√£o da mascara para detec√ß√£o de bordas na imagem com blur
edgesx_2 = cv2.Scharr(blurred_2, -1, dx=1, dy=0, scale=1,
                       delta=0, borderType=cv2.BORDER_DEFAULT)
edgesy_2 = cv2.Scharr(blurred_2, -1, dx=0, dy=1, scale=0.75,
                       delta=0, borderType=cv2.BORDER_DEFAULT)
edges2_2 = edgesx_2 + edgesy_2

# Print do resultado do processamento
cv2_imshow(edges2_2)
```
